```python
# import modules:
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
print("All modules successfully imported!")
```

```python
#import covid 19 corona_dataset
corona_dataset = pd.read_csv('C:\\Users\\asus\\Desktop\\covid19.csv')
corona_dataset.head()
```

```python
corona_dataset.shape
```

```python
#drop useless columns
corona_dataset.drop(["Lat","Long"],axis = 1,inplace = True)
corona_dataset.head()
```

```python
#Aggregating the rows by the country
df = corona_dataset.groupby("Country/Region").sum()
df.head(10)
```

```python
df.shape
```

```python
#Visualizing data related to a country
df.loc["China"].plot()
df.loc["India"].plot()
df.loc["Russia"].plot()
plt.legend()
```

```python
df.loc['India'][70:75].plot()
```

```python
df.loc['India'].diff().plot()
```

```python
df.loc['India'].diff().max()
```

```python
df.loc['Russia'].diff().max()
```

```python
countries = list(df.index)
max_infection = []
for corona in countries:
    max_infection.append(df.loc[corona].diff().max())
df['max_infections'] = max_infection 
df.head()
```

```python
#new columns
new_columns = pd.DataFrame(df["max_infections"])
new_columns.head()
```

```python
happiness_dataset = pd.read_csv("C:\\Users\\asus\\Desktop\\happiness.csv")
happiness_dataset.head()
```
